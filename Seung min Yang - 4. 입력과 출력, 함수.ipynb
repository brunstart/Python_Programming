### 1. 세 개의 양의 정수를 입력받아 그 합이 짝수이면 가장 큰 수를 출력하고, 홀수이면 가장 작은 수를 출력하는 프로그램을 작성하시오.

```
def func1():
  a = int(input('첫 번째 양의 정수 입력: '))
  b = int(input('두 번째 양의 정수 입력: '))
  c = int(input('세 번째 양의 정수 입력: '))

  if (a + b + c) % 2 == 0:
    print("합: 짝수, max =",max(a,b,c))
  else:
    print("합: 홀수, min =",min(a,b,c))
    
func1()
```

### 2. 두 점 사이의 거리를 구하는 프로그램을 작성하시오.


*   두 점의 x, y 좌표를 각각 입력 받는다.
*   두 점 사이의 거리를 구하는 공식은 다음과 같다.
$${\sqrt{(x1-x2)^2 + (y1-y2)^2}}$$

*   제곱근은 (~)**0.5 로 표현한다.





```
def func2():
  x1, y1 = map(int,input('x 좌표(x1,y1) 입력: ').split())
  x2, y2 = map(int,input('y 좌표(x2,y2) 입력: ').split())
  
  return ((x1-x2)**2 + (y1 - y2)**2)**0.5

func2()
```

### 3. 2~99 사이의 정수를 입력하면 그 수의 약수를 출력하는 프로그램을 작성하시오.

```
def func3():

  n = int(input('2 ~ 99 사이의 정수를 입력하시오 : '))

  if n < 2 or n > 99:
    print("범위를 벗어난 수 입력")
    return
  print(n,'의 약수 :')  
  for i in range(1,n+1):
    if n % i == 0:
      print(i)

func3()
```

### 5. 리스트로 작성한 학생 점수를 입력하면 최고점을 구하는 함수 high_score를 정의하고, 이 함수를 이용하여 아래 점수 중 최고점을 구하는 프로그램을 작성하시오.

score = [60, 72, 98, 48, 100, 86]

```
score = [60,72,98,48,100,86]

def high_score(li):
  print('최고점 :',max(li),'점')

high_score(score)
```
